## IMPORTING THE LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

## IMPORTING AND ANALYSIS THE DATASET

```python
dataset = pd.read_csv('creditcard.csv')
```

```python
print(dataset.info())
```

```python
print(dataset.shape)
```

```python
dataset.isnull().sum()                                  #early info of the dataset
```

```python
fraud = dataset[dataset['Class'] == 1]     #all rows with dataset['Class'] = 1
valid = dataset[dataset['Class'] == 0]     #all rows with dataset['Class'] = 0
outlierFraction = (len(fraud))/float(len(valid)) 
print(outlierFraction) 
print('Fraud Cases: {}'.format(len(dataset[dataset['Class'] == 1]))) 
print('Valid Transactions: {}'.format(len(dataset[dataset['Class'] == 0]))) 
```

```python
print('Amount details of the fraudulent transaction') 
fraud.Amount.describe() 
```

```python
print('details of valid transaction') 
valid.Amount.describe() 
```

```python
corrmat = dataset.corr() 
fig = plt.figure(figsize = (12, 12)) 
sb.heatmap(corrmat, vmax = .8, square = True) 
plt.show() 
```

## SPLITTING TRAINING SET AND TEST SET

```python
features = dataset.drop(['Class'], axis=1)
target = dataset['Class']

print(features.shape)
print(target.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## PREDICTING TEST SET

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score

y_pred = classifier.predict(X_test)
# print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
print('Training Accuracy : ', metrics.roc_auc_score(y_train, classifier.predict(X_train)))
print('Validation Accuracy : ', metrics.roc_auc_score(y_test, classifier.predict(X_test)))
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Correct: ", cm[0][0] + cm[1][1])
print("Incorrect: ", cm[0][1] + cm[1][0])
```

```python
# Evaluating the classifier 
# printing every score of the classifier 
# scoring in anything 
from sklearn.metrics import classification_report, accuracy_score 
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import f1_score, matthews_corrcoef 
from sklearn.metrics import confusion_matrix 

n_outliers = len(fraud) 
n_errors = (y_pred != y_test).sum() 
print(f"No of incorrect: {n_errors}")

acc = accuracy_score(y_test, y_pred) 
print("The accuracy is {}".format(acc)) 

prec = precision_score(y_test, y_pred) 
print("The precision is {}".format(prec))  #predicted fraud was actually fraud

rec = recall_score(y_test, y_pred)         #actual fraud was predicted fraud
print("The recall is {}".format(rec)) 

f1 = f1_score(y_test, y_pred) 
print("The F1-Score is {}".format(f1)) 

MCC = matthews_corrcoef(y_test, y_pred) 
print("The Matthews correlation coefficient is {}".format(MCC)) 
```

```python
LABELS = ['Normal', 'Fraud'] 
conf_matrix = confusion_matrix(y_test, y_pred) 
plt.figure(figsize =(7, 7)) 
sb.heatmap(conf_matrix, xticklabels = LABELS,  
            yticklabels = LABELS, annot = True, fmt ="d"); 
plt.title("Confusion matrix") 
plt.ylabel('True class') 
plt.xlabel('Predicted class') 
plt.show() 
```

## Applying k-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 3)
print("Max accuracy: {:.5f} %".format(max(accuracies)*100))
print("Min accuracy: {:.5f} %".format(min(accuracies)*100))
print("Accuracy: {:.5f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.5f} %".format(accuracies.std()*100))
```
